```python
from sklearn import tree
import pandas as pd
import os
import numpy as np
```

```python
dfraw = pd.read_csv(os.path.join("ML_minbreweriesexpectedpercounty.csv"))


booleanregionalcount = []
for item in dfraw["countregional"]:
    if item != 0:
        booleanregionalcount.append(1)
    elif item == 0:
        booleanregionalcount.append(0)
        
dfraw['booleanregionalcount'] = booleanregionalcount
dfraw
```

```python
dfyes = dfraw[dfraw["booleanregionalcount"] == 1]
dfno = dfraw[dfraw["booleanregionalcount"] == 0]

dfnosample = dfno.sample(10*len(dfyes))
dfnosample

new_df = pd.concat([dfyes,dfnosample],ignore_index=True,copy=True)
new_df
```

```python
target = new_df['booleanregionalcount']
target_names = ['no','yes']
target
```

```python
data = new_df.copy(deep=True)[['median_HHI_2019', 'bach_deg_pct_14to18','population2019', 'whitepercent', 'blackpercent',
       'hispanicpercent', 'millennialpercent', 'genxpercent', 'boomerpercent','densitypermile', 'popchange', 'malepercent']]
data = data.replace([np.inf, -np.inf], np.nan).fillna(data.mean())
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
# rf = RandomForestRegressor(n_estimators=200)
# rf = rf.fit(X_train, y_train)
# rf.score(X_test, y_test)

pipe = make_pipeline(StandardScaler(), RandomForestClassifier(n_estimators=1000, max_depth=6,n_jobs=-1))
rf = pipe.fit(X_train, y_train)  # apply scaling on training data

print(rf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

```python
fulldata = dfraw.copy(deep=True)[['median_HHI_2019', 'bach_deg_pct_14to18','population2019', 'whitepercent', 'blackpercent',
       'hispanicpercent', 'millennialpercent', 'genxpercent', 'boomerpercent','densitypermile', 'popchange', 'malepercent']]
fulldata = fulldata.replace([np.inf, -np.inf], np.nan).fillna(data.mean())
dfraw['expectedregional'] = rf.predict(fulldata)
dfraw
```
