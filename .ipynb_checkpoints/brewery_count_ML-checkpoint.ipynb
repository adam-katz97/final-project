```python
from sklearn import tree
import pandas as pd
import os
import numpy as np
```

```python
dfraw = pd.read_csv(os.path.join("rg_proj2_resources/countydemographicsFINALPROJwithbrewcount.csv"))
dfzero = dfraw[dfraw["breweriesTOTALper10000"] == 0]
dfnonzero = dfraw[dfraw["breweriesTOTALper10000"] != 0]
dfnonzero
```

```python
working_target = dfnonzero["breweriesTOTALper10000"]
from sklearn.cluster import KMeans
target = pd.qcut(working_target,2,labels=["low","high"])
target
```

```python
pd.value_counts(target)
```

```python
dfnonzero['target_value'] = target
dfzero['target_value'] = "zero"
```

```python
new_df = pd.concat([dfnonzero,dfzero.sample(n=642)],ignore_index=True,copy=True)
new_df
```

```python
y = new_df['target_value']
```

```python
data = new_df.copy(deep=True)[['median_HHI_2019', 'bach_deg_pct_14to18','population2019', 'whitepercent', 'blackpercent',
       'hispanicpercent', 'millennialpercent', 'genxpercent', 'boomerpercent','densitypermile', 'popchange', 'malepercent']]
data = data.replace([np.inf, -np.inf], np.nan).fillna(data.mean())
data
```

```python
data = data.astype('float32')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, y, random_state=42)
print(X_train)
print(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
# rf = RandomForestRegressor(n_estimators=200)
# rf = rf.fit(X_train, y_train)
# rf.score(X_test, y_test)

pipe = make_pipeline(StandardScaler(), RandomForestClassifier(n_estimators=1000, max_depth=6,n_jobs=-1))
rf = pipe.fit(X_train, y_train)  # apply scaling on training data

print(rf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
```
